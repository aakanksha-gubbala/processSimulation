<p style="font-family:futura;">Experiment: Flow Through Pipes</p>
The aim of the experiment is to determine the effect of Reynolds number on fanning friction factor for different flow regimes in a smooth pipe ($\epsilon / D \approx 0$). The experiment is performed for two process fluids - water and kerosene.

<img src="pfd.png" style="float: center;" width=500 height=500/>

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import scipy.optimize as opt
from scipy.constants import g
from sklearn.metrics import r2_score

%config InlineBackend.figure_format = 'retina'

style.use("classic")
````

<p style="font-family:futura;">Formulae</p>

$$\Delta P = \Delta H_m (\rho_m - \rho_f) g$$

$$f_{experiment} = \frac{D \Delta P}{2 \rho v^2 L}$$

For laminar regime:

$$f = \frac{16}{Re}$$

The Blasius equation for turbulent regime is

$$f = 0.079 Re^{-0.25}$$

The von Kármán-Nikuradse Correlation for turbulent regime is

$$\frac{1}{\sqrt{f}} = 4\log{(Re\sqrt{f})} - 0.4$$

---
The properties of both fluids are specified. We have to fit a experimental data of $f$ anf $Re$ to the equation type $f = A \times Re^B$. We also need to calculate the value friction factor using correlations.

````python
FluidProperties = {'Water': [1000, 8.5e-4],
                   'Kerosene': [820, 2.15e-3]}

def correlation_f_Re(df):
    f = df['Friction Factor']
    Re = df['Reynolds Number']
    if Re < 2100:
        f_corr = 16 / Re
    else:
        func = lambda x: 4 * np.log(Re * np.sqrt(x)) - 0.4 - 1 / np.sqrt(x)
        x0 = 0.001
        f_corr = opt.fsolve(func, x0, maxfev=1000000)[0]
    return f_corr


def f_Re(x, A, B):
    return A * x ** B
````

---
The class ```FlowThroughPipes``` reads a sheet from the excel file ```"FlowThroughPipes_DATA.xlsx"```. The user has to input the sheet-name, diameter and length of pipe. Each sheet contains a single dataset. The input data file looks like this:
<img src="input_excel.png" style="float: center;" width=500 height=500/>

````python
class FlowThroughPipes:
    def __init__(self, sheetname, D, L, fluid):
        self.df = pd.read_excel("FlowThroughPipes_DATA.xlsx", sheet_name=sheetname)
        self.D = D
        self.L = L
        self.fluid = fluid

    def CalculationTable(self):
        self.df.index += 1
        ManometerDensity = self.df['Manometer Fluid'].replace('Mercury', 13600.0).replace('CCl4', 1600.0)

        rho_f = FluidProperties[self.fluid][0]
        mu_f = FluidProperties[self.fluid][1]

        dHm = self.df['dHm (cm)']
        self.df['Delta P (Pa)'] = (dHm * 0.01) * (ManometerDensity - rho_f) * g

        Q = self.df['Q (LPM)'] * 0.001 / 60
        self.df['Velocity (m/s)'] = 4 * Q / (np.pi * self.D ** 2)

        self.df['Reynolds Number'] = self.D * self.df['Velocity (m/s)'] * rho_f / mu_f

        self.df['Friction Factor'] = self.df['Delta P (Pa)'] * self.D / (
                2 * rho_f * self.df['Velocity (m/s)'] ** 2 * self.L)

        self.df['f (correlations)'] = self.df.apply(correlation_f_Re, axis=1)

        self.df['Flow Regime'] = self.df['Reynolds Number'].map(lambda x: 'Laminar' if x < 2000 else 'Turbulent')

        return self.df.round({'Reynolds Number': 0, 'Velocity (m/s)': 3, 'Delta P (Pa)': 2})

    def Graphs(self):
        plt.figure(facecolor='white')
        plt.title(r"$f$ vs. $Re$ (%s)" % self.fluid, fontsize=18)
        plt.grid()
        plt.xlim(0, 1.1 * max(self.df['Reynolds Number']))
        plt.ylim(0, 1.1 * max(self.df['Friction Factor']))
        plt.xlabel(r"$Re$")
        plt.ylabel(r"$f$")
        try:
            lam = self.df[self.df['Flow Regime'] == 'Laminar']
            [A_lam, B_lam], params = opt.curve_fit(f_Re, lam['Reynolds Number'], lam['Friction Factor'])
            print("Laminar FLow: f = %0.4f Re^(%0.3f)" % (A_lam, B_lam))
            print("R2 = %0.4f\n" % r2_score(lam['Friction Factor'], f_Re(lam['Reynolds Number'], A_lam, B_lam)))

            plt.plot(lam['Reynolds Number'], f_Re(lam['Reynolds Number'], A_lam, B_lam),
                     color='blue', linestyle='--', label='Curve Fitting (laminar)')
            plt.scatter(lam['Reynolds Number'], lam['Friction Factor'],
                        color='blue', label='Experiment (laminar)')

            tur = self.df[self.df['Flow Regime'] == 'Turbulent']
            [A_tur, B_tur], params = opt.curve_fit(f_Re, tur['Reynolds Number'], tur['Friction Factor'])
            print("Turbulent Flow: f = %0.4f Re^(%0.3f)" % (A_tur, B_tur))
            print("R2 = %0.4f\n" % r2_score(tur['Friction Factor'], f_Re(tur['Reynolds Number'], A_tur, B_tur)))

            plt.plot(tur['Reynolds Number'], f_Re(tur['Reynolds Number'], A_tur, B_tur),
                     color='green', linestyle='--', label='Curve Fitting (turbulent)')
            plt.scatter(tur['Reynolds Number'], tur['Friction Factor'],
                        color='green', label='Experiment (turbulent)')
            plt.legend(bbox_to_anchor=(1.26, 1), fontsize=6)

        except:
            tur = self.df[self.df['Flow Regime'] == 'Turbulent']
            [A_tur, B_tur], params = opt.curve_fit(f_Re, tur['Reynolds Number'], tur['Friction Factor'])
            print("Turbulent Flow: f = %0.4f Re^(%0.3f)" % (A_tur, B_tur))
            print("R2 = %0.4f\n" % r2_score(tur['Friction Factor'], f_Re(tur['Reynolds Number'], A_tur, B_tur)))

            plt.plot(tur['Reynolds Number'], f_Re(tur['Reynolds Number'], A_tur, B_tur),
                     color='green', linestyle='--', label='Curve Fitting (turbulent)')
            plt.scatter(tur['Reynolds Number'], tur['Friction Factor'],
                        color='green', label='Experiment (turbulent)')
            plt.legend(bbox_to_anchor=(1.26, 1), fontsize=6)
````
