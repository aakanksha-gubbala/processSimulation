<p style="font-family:futura;">Dynamics of a Heated Tank with Proportional/Integral (PI) Temperature Control</p>

<img src="pfd.png" alt="Markdown Monster icon" style="float: center;" width=500 height=500/>

<p style="font-family:futura;">Modeling and Control Equations</p>

$$\frac{dT}{dt} = \frac{W C_p (T_i - T) + q}{\rho V C_p},\ T(t = 0) = T_r$$

$$\frac{dT_0}{dt} = \big[T - T_0 - \big(\frac{\tau_d}{2}\big)\big(\frac{dT}{dt}\big)\big]\frac{2}{\tau_d},\ T_0(t = 0) = T_r$$

$$\frac{dT_m}{dt} = \frac{T_0 - T_m}{\tau_m},\ T_m(t = 0) = T_r$$

$$q = q_s + K_c (T_r - T_m) + \frac{K_c}{\tau_I}(errsum)$$

where

$$q_s = WC_p (T_r - T_{is})$$

and

$$\frac{d(errsum)}{dt} = T_r - T_m,\ errsum(t = 0) = 0$$

```python
import numpy as np
from scipy.constants import R
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
from scipy.integrate import odeint

from IPython.display import display, Math

%config InlineBackend.figure_format = 'retina'

style.use("classic")
```

```python
# Baseline System and ConT_rol Parameters
rhoVCp = 4000  # kJ/˚C
WCp = 500  # kJ/min/˚C
Tis = 60  # ˚C
Tr = 80  # ˚C
tau_d = 1  # min
tau_m = 5  # min
tau_I = 2  # min
```

```python
def ODE_System(x, t, Kc):
    
    ### The ODEs needed to be solved are defined in this function

    [T, T0, Tm, errsum] = x
    
    # The system is initially operating at design steady state at a temperature of 80˚C, 
    # and inlet temperature Ti is suddenly changed to 40˚C at time t = 10 min.
    
    if t < 10:
        Ti = 60  # ˚C
    else:
        Ti = 40  # ˚C

    qs = WCp * (Tr - Tis)
    q = qs + Kc * (Tr - Tm) + (Kc / tau_I) * errsum

    dTdt = (WCp * (Ti - T) + q) / rhoVCp
    dT0dt = (T - T0 - (tau_d / 2) * (dTdt)) * (2 / tau_d)
    dTmdt = (T0 - Tm) / tau_m
    derrsumdt = Tr - Tm

    return [dTdt, dT0dt, dTmdt, derrsumdt]
```

```python
x0 = [80, 80, 80, 0]
Kc = [0, 50, 500, 5000]  # kJ/min/˚C

# defining time arrays
t_0 = np.linspace(0, 60, 500)
t_50 = np.linspace(0, 120, 500)
t_500 = np.linspace(0, 180, 500)

# sol = [T, T0, Tm, errsum]
sol_0 = odeint(ODE_System, x0, t_0, args=(Kc[0],))
sol_50 = odeint(ODE_System, x0, t_50, args=(Kc[1],))
sol_500 = odeint(ODE_System, x0, t_500, args=(Kc[2],))


def plot_sol(t, sol, Kc):
    fig = plt.figure(facecolor="white")
    plt.suptitle(r"$K_c = %d\ \frac{kJ}{min.K}$" % Kc, fontsize=18)
    plt.grid()
    plt.plot(t, sol[:, 0], label=r'$T$', color='#00cc44', linewidth=1.5)
    plt.plot(t, sol[:, 1], label=r'$T_0$', color='#ff471a', linewidth=1.5)
    plt.plot(t, sol[:, 2], label=r'$T_m$', color='#3377ff', linewidth=1.5)
    plt.xlabel("Time (min)")
    plt.ylabel("Temperature (˚C)")
    plt.legend(loc="best")
```

```python
plot_sol(t_0, sol_0, Kc[0])
```

```python
plot_sol(t_50, sol_50, Kc[1])
```

```python
plot_sol(t_500, sol_500, Kc[2])
```

```python
# There are limits on 'q' in the modified version of the ODE
def ODE_System_Modified(x, t, Kc):
    [T, T0, Tm, errsum] = x

    Ti = Tis

    if t < 10:
        Tr = 80  # ̊C
    else:
        Tr = 90  # ̊C

    qs = WCp * (Tr - Tis)
    q = qs + Kc * (Tr - Tm)

    if q < 0:
        q = 0
    else:
        if q >= 2.6 * qs:
            q = 2.6 * qs
        else:
            q = q

    dTdt = (WCp * (Ti - T) + q) / rhoVCp
    dT0dt = (T - T0 - (tau_d / 2) * (dTdt)) * (2 / tau_d)
    dTmdt = (T0 - Tm) / tau_m
    derrsumdt = Tr - Tm

    return [dTdt, dT0dt, dTmdt, derrsumdt]

t = np.linspace(0, 180, 1000)
sol = odeint(ODE_System_Modified, x0, t, args=(Kc[3],))
plot_sol(t, sol, Kc[3])
```
